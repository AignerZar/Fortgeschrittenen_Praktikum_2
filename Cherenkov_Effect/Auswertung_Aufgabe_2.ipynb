### For that exercise, also the time differences of channel 1 and channel 2 need to be calculated, therefore the file "Compute_time_difference.py", in the excel file "Data_exercise_2" the obtained data can be found.

```python
d = 1.662 # distance between tanks [m]
d_error = 0.005 # error of distance [m]

t_ns = np.array([3.881, 3.019, 2.974, 4.052]) # time differences [ns]
t_ns_err = np.array([0.025, 0.025, 0.025, 0.025]) # error time differences [ns]

t = t_ns * 1e-9 # time in [s]
t_err = t_ns_err * 1e-9 # error time in [s]

# computing the velocity
v = d / t
v_err = np.sqrt( (1/t * d_error)**2 + (d / t**2 * t_err)**2 ) # gaussian error propagation

c = 3e8  # speed of light
v_c = v / c
v_c_err = v_err / c


for i in range(len(v)):
    print(f"v_{i+1} = {v[i]:.2e} +- {v_err[i]:.2e} m/s")

for i in range(len(v_c)):
    print(f"v_{i+1} in c = {v_c[i]:.2e} +- {v_c_err[i]:.2e} c")
```
